* (10 points) จากข้อมูลที่ให้มา เตรียมข้อมูลให้พร้อมโดยการจัดการกับค่าว่างในข้อมูล ให้อธิบายวิธีที่ท่านเลือกใช้ พร้อมยกเหตุผลประกอบ

```python
# จากข้อมูลที่ให้มา เตรียมข้อมูลให้พร้อมโดยการจัดการกับค่าว่างในข้อมูล ให้อธิบายวิธีที่ท่านเลือกใช้ พร้อมยกเหตุผลประกอบ

#  จัดการกับค่าว่างในข้อมูล
#  สำหรับตัวแปร Age จะใช้ค่าเฉลี่ย (mean) แทนค่าว่าง เนื่องจากเป็นตัวแปรเชิงปริมาณ และการใช้ค่าเฉลี่ยเป็นวิธีที่ง่ายและรวดเร็วในการเติมค่าว่าง
df_titanic['age'].fillna(df_titanic['age'].mean(), inplace=True)

#  สำหรับตัวแปร Embarked จะใช้ค่าที่พบมากที่สุด (mode) แทนค่าว่าง เนื่องจากเป็นตัวแปรเชิงคุณภาพ และการใช้ค่าที่พบมากที่สุดเป็นวิธีที่เหมาะสมในการเติมค่าว่าง
df_titanic['embarked'].fillna(df_titanic['embarked'].mode()[0], inplace=True)

#  สำหรับตัวแปร cabin จะลบแถวที่มีค่าว่าง เนื่องจากเป็นตัวแปรเชิงคุณภาพ และการลบแถวที่มีค่าว่างเป็นวิธีที่เหมาะสมที่สุดในกรณีนี้
df_titanic.dropna(subset=['cabin'], inplace=True)

#  ตรวจสอบค่าว่างอีกครั้ง
missing_values = df_titanic.isnull().sum()
print(missing_values)
```

* แปลงข้อมูลเชิงคุณภาพให้เป็นตัวเลข \
    i. (10 points) ใช้ one-hot encoding ในการแปลงตัวแปร Embarked \
    ii. (10 points) แปลงตัวแปร Sex โดยให้เพศชายเป็น 0 และเพศหญิงเป็น 1 \
    iii. (10 points) สร้างตัวแปร FamilySize ซึ่งเป็นผลรวมของตัวแปร SibSp (number of siblings/spouses aboard) และ Parch (number of parents/children aboard) \
    iv. (10 points) สร้างตัวแปร IsAlone ซึ่งมีค่าเป็น 1 ถ้า FamilySize เป็น 0, และมีค่าเป็นศูนย์เมื่อเป็นอื่น

```python


# i. ใช้ one-hot encoding ในการแปลงตัวแปร Embarked
df_titanic = pd.get_dummies(df_titanic, columns=['embarked'], prefix=['embarked'])

# ii. แปลงตัวแปร Sex โดยให้เพศชายเป็น 0 และเพศหญิงเป็น 1
df_titanic['sex'] = df_titanic['sex'].map({'male': 0, 'female': 1})

# iii. สร้างตัวแปร FamilySize ซึ่งเป็นผลรวมของตัวแปร SibSp และ Parch
df_titanic['familysize'] = df_titanic['sibsp'] + df_titanic['parch']

# iv. สร้างตัวแปร IsAlone ซึ่งมีค่าเป็น 1 ถ้า FamilySize เป็น 0, และมีค่าเป็น 0 เมื่อเป็นอื่น
df_titanic['isalone'] = np.where(df_titanic['familysize'] == 0, 1, 0)
```

* ทำการ Normalize ข้อมูล \
    i. (10 points) ตัวแปร Fare และ Age ใช้ Min-Max Scaling \
    ii. (10 points) สร้างตัวแปรใหม่ Fare_Per_Person ด้วยการนำตัวแปร Fare มาหารด้วย FamilySize + 1 \
    iii. (10 points) สร้างตัวแปร AgeGroup ที่ระบุกลุ่มอายุของตัวอย่าง โดยแบ่งเป็นกลุ่มดังนี้ [0, 12), [12, 18), [18, 50) และ [50, 100)

```python
import pandas as pd
import numpy as np
pd.cut(np.array([1, 7, 5, 4, 6, 3]), 3)
```

```python
pd.cut(np.array([1, 7, 5, 4, 6, 3]), [0,2,5,8])
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# i. ตัวแปร Fare และ Age ใช้ Min-Max Scaling
scaler = MinMaxScaler()
df_titanic[['fare', 'age']] = scaler.fit_transform(df_titanic[['fare', 'age']])

# ii. สร้างตัวแปรใหม่ Fare_Per_Person ด้วยการนำตัวแปร Fare มาหารด้วย FamilySize + 1
df_titanic['fare_per_person'] = df_titanic['fare'] / (df_titanic['familysize'] + 1)

# iii. สร้างตัวแปร AgeGroup ที่ระบุกลุ่มอายุของตัวอย่าง โดยแบ่งเป็นกลุ่มดังนี้ [0, 12), [12, 18), [18, 50) และ [50, 100)
bins = [0, 12, 18, 50, 100]
labels = ['Child', 'Teenager', 'Adult', 'Senior']
df_titanic['agegroup'] = pd.cut(df_titanic['age'] * 100, bins=bins, labels=labels, include_lowest=True)
```

* (10 points) นำเสนอการแจกแจงหรือทำความเข้าใจตัวแปรสำคัญในข้อมูล เช่น Age, Fare, and Pclass ด้วยการสร้างฮิสโตแกรมหรือแผนภาพกล่อง

```python
import matplotlib.pyplot as plt
# Histograms for Age, Fare, and Pclass
df_titanic['age'].hist()
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

df_titanic['fare'].hist()
plt.title('Histogram of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()

df_titanic['pclass'].hist()
plt.title('Histogram of Pclass')
plt.xlabel('Pclass')
plt.ylabel('Frequency')
plt.show()

# Box plots for Age, Fare, and Pclass
df_titanic.boxplot(column='age')
plt.title('Boxplot of Age')
plt.show()

df_titanic.boxplot(column='fare')
plt.title('Boxplot of Fare')
plt.show()

df_titanic.boxplot(column='pclass')
plt.title('Boxplot of Pclass')
plt.show()
```

* เมื่อเตรียมข้อมูลพร้อมแล้ว ให้หาความสัมพันธ์ของตัวแปรต่อไปนี้จากข้อมูล ซึ่งอาจนำเสนอด้วยสถิติและแผนภูมิหรือกราฟก็ได้: \
    i. (10 points) survived และ pclass \
    ii. (10 points) survived และ sex \
    iii. (10 points) survived และ age \
    iv. (10 points) จำนวนสมาชิกครอบครัวที่รอดชีวิต และ Fare

```python
import pandas as pd
import matplotlib.pyplot as plt
# i. survived และ pclass
print(pd.crosstab(df_titanic['survived'], df_titanic['pclass']))
df_titanic.groupby('pclass')['survived'].mean().plot(kind='bar')
plt.title('Survival Rate by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Survival Rate')
plt.show()

# ii. survived และ sex
print(pd.crosstab(df_titanic['survived'], df_titanic['sex']))
df_titanic.groupby('sex')['survived'].mean().plot(kind='bar')
plt.title('Survival Rate by Sex')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.show()

# iii. survived และ age
plt.scatter(df_titanic['age'], df_titanic['survived'])
plt.title('Survival Rate by Age')
plt.xlabel('Age')
plt.ylabel('Survived')
plt.show()

# iv. จำนวนสมาชิกครอบครัวที่รอดชีวิต และ Fare
plt.scatter(df_titanic['familysize'], df_titanic['fare'])
plt.title('Fare vs. Family Size')
plt.xlabel('Family Size')
plt.ylabel('Fare')
plt.show()
```

* (10 points) สร้างตารางสรุปข้อมูลเพื่อหาคุณลักษณะของผู้โดยสารเรือไททานิค และศึกษาเกี่ยวกับการเสียชีวิตของผู้โดยสาร ทำการสรุปผลที่ได้จากตาราง

```python
# สร้างตารางสรุปข้อมูล
summary_table = df_titanic.groupby(['survived', 'sex', 'pclass'])['age'].agg(['mean', 'count'])

# แสดงตารางสรุปข้อมูล
print(summary_table)

# สรุปผลที่ได้จากตาราง
# จากตารางสรุปข้อมูล พบว่าผู้หญิงมีอัตราการรอดชีวิตสูงกว่าผู้ชาย
# ผู้โดยสารชั้น 1 มีอัตราการรอดชีวิตสูงกว่าผู้โดยสารชั้น 2 และ 3
# อายุเฉลี่ยของผู้โดยสารที่รอดชีวิตมีแนวโน้มสูงกว่าผู้โดยสารที่เสียชีวิต
```
